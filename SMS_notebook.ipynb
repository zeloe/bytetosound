```python
def spec(y,fftsize):
  D = lr.stft(y,n_fft = fftsize)  # STFT of y
  S_db = lr.amplitude_to_db(np.abs(D), ref=np.max)
  plt.figure()
  lr.display.specshow(S_db)
  plt.colorbar()
```
